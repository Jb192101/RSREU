```python
from functools import lru_cache
import itertools
from functools import partial
from itertools import product
```

```python
# Задание 1
# Проверить, есть ли чётные числа
def has_even_number(numbers):
    return any(num % 2 for num in numbers)

# Сортировка
def sort_key(student):
    name, score = student
    return (score)

# Кэширование результатов
@lru_cache(maxsize=128)
def factorial(n):
    prod = 1
    for i in range(n):
        prod *= i
    
    return prod

# partial
def power(base, exponent):
    return base ** exponent
```

```python
print("Задание 1")
print(has_even_number([1, 2, 3, 4, 5]))

print()
print("Задание 2")
students = [("Алиса", 85), ("Боб", 75), ("Чарли", 95), ("Дэвид", 75)]
students.sort(key=sort_key)
print(students)

print()
print("Задание 3")
print(factorial(5))
print(factorial(6))
print(factorial(5))
print(factorial.cache_info())

print()
print("Задание 4")
power_to_2 = partial(power, 2)
print(power_to_2(3))
print(power_to_2(4))

print()
print("Задание 5")
transactions = [{"date": "2023-01-15", "amount": 100}, {"date": "2023-01-16", "amount": -50}, {"date": "2023-02-01", "amount": 200}, {"date": "2023-02-15", "amount": -75}]
result_iterator = itertools.dropwhile(lambda x: x["amount"] > 0, transactions)
result = list(result_iterator)

print("Транзакции, начиная с первой отрицательной:")
for transaction in result:
    print(f"Дата: {transaction['date']}, Сумма: {transaction['amount']}")

print()
print("Задание 6")
daily_sales = [10, 15, 7, 20, 13]
cumulative_sum = list(itertools.accumulate(daily_sales))
print("Накопительная сумма продаж по дням:")
print(cumulative_sum)
```

```python
hex_digits = '0123456789ABCDEF'
is_even = lambda d: d in '02468ACE'

condition = lambda num: (
    num[0] != '0' and
    len(set(num)) == 5 and
    all(is_even(num[i]) != is_even(num[i+1]) for i in range(4))
)

valid_numbers = list(filter(condition, map(''.join, product(hex_digits, repeat=5))))

print(f"Количество чисел: {len(valid_numbers)}")
```
